### Removing of columns from DataFrame 

```python
# list of prefixes for removing
prefix = ['Time from Start to Finish (seconds)','Q2_OTHER_TEXT',
          'Q5_OTHER_TEXT', 'Q6', 'Q7', 'Q8',
          'Q9_','Q11', 'Q12_','Q14_Part','Q13_OTHER','Q14_OTHER_', 'Q16_OTHER_',
          'Q17_OTHER_', 'Q18_OTHER_', 'Q19','Q19_', 'Q19_OTHER', 
          'Q20_OTHER','Q21_Part',
          'Q21_OTHER','Q22', 'Q24_Part', 'Q24_OTHER_TEXT',
          'Q25_Part', 'Q25_OTHER', 'Q26_Part', 'Q26_OTHER',
          'Q27_Part', 'Q27_OTHER','Q28_OTHER', 'Q29_Part','Q29_OTHER', 
          'Q30_Part', 'Q30_OTHER', 'Q31_Part', 'Q31_OTHER',
          'Q32_Part', 'Q32_OTHER', 'Q33_Part', 'Q33_OTHER', 'Q34_Part',
          'Q34_OTHER']

# find columns with prefix
columns_to_drop = [
     col for col in df19.columns if any(col.startswith(p) for p in prefix)
     ]
# print(columns_to_drop)

# remove columns from dataframe
df19 = df19.drop(columns=columns_to_drop)

# new dataframe
df19.shape

df19.columns

#columns in new dataframe
for col in df19.columns:
     print(col)
```

### Correction of header

Combination of two rows for preparation of main header

```python
#original header
original_cols = df19.columns.tolist()
line_0 = df19.iloc[0].tolist()
# print(original_cols)
# print(line_0)

# new header combination with first one
new_header = [f"{col}_{val}" for col, val in zip(original_cols, line_0)]
# print(new_header)

# setting new header and remove first row
df19.columns = new_header
df19 = df19.iloc[1:].reset_index(drop=True)
df19
```

### Splitting columns that contain multiple tools or their versions

Split column Ide's_Visual_Studio/VisualStudio_Code into two columns

```python
col_index = df19.columns.get_loc("Ide's_Visual_Studio/VisualStudio_Code")

# Split column
new_cols = df19["Ide's_Visual_Studio/VisualStudio_Code"].str.split('/', expand=True)
new_cols.columns = ["Ide's_Visual_Studio", "Ide's_Visual_Studio_Code"]  #new names

# Drop original column
df19.drop(columns="Ide's_Visual_Studio/VisualStudio_Code", inplace=True)

# Insert new columns to original position
for i, col_name in enumerate(new_cols.columns):
    df19.insert(col_index + i, col_name, new_cols[col_name])

df19.shape
```
