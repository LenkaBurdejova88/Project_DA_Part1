### Removing of columns from DataFrame 

```python
# list of prefixes for removing
prefix = ['Time from Start to Finish (seconds)', 'Q8','Q11', 'Q12_', 'Q13',
          'Q17_','Q18_', 'Q19_','Q20', 'Q21', 'Q22', 'Q23', 'Q25', 'Q26_',
          'Q27_', 'Q28_','Q29_', 'Q30', 'Q30_', 'Q31', 'Q31_', 'Q32', 'Q32_',
          'Q33_', 'Q33_B','Q34_', 'Q35_', 'Q36_', 'Q39_']

# find columns with prefix
columns_to_drop = [
     col for col in df20.columns if any(col.startswith(p) for p in prefix)
     ]

# print(columns_to_drop)

# remove columns from dataframe
df20 = df20.drop(columns=columns_to_drop)

# new dataframe
# df20.shape

for col in df20.columns: #columns in new dataframe as list
     print(col)
```

### Correction of headers

Combination of two rows for preparation of main header

```python
#original header
original_cols = df20.columns.tolist()
line_0 = df20.iloc[0].tolist()

# new header combination with first one
new_header = [f"{col}_{val}" for col, val in zip(original_cols, line_0)]

# setting new header and remove first row
df20.columns = new_header
df20 = df20.iloc[1:].reset_index(drop=True)
df20
```

### Splitting columns that contain multiple tools or their versions

Split column Ide's_Vim/Emacs into two columns

```python
col_index = df20.columns.get_loc("Ide's_Vim/Emacs")

# Split column
new_cols = df20["Ide's_Vim/Emacs"].str.split('/', expand=True)
new_cols.columns = ["Ide's_Vim", "Ide's_Emacs"]  

# Drop original column
df20.drop(columns="Ide's_Vim/Emacs", inplace=True)

# Insert new column to same position
for i, col_name in enumerate(new_cols.columns):
    df20.insert(col_index + i, col_name, new_cols[col_name])

df20.shape
```
