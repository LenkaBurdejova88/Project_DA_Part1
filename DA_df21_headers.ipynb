### Fix encoding

```python
import pandas as pd

df21 = pd.read_csv("2021_kaggle_ds_and_ml_survey_responses_only.csv", low_memory=False)

def fix_encoding(text):
    if isinstance(text, str):
        try:
            return text.encode("latin1").decode("utf-8")
        except UnicodeEncodeError:
            return text
    return text

df21 = df21.apply(lambda col: col.map(fix_encoding) if col.dtype == 'object' else col)

df21
```

### Removing of columns from DataFrame 

```python
# list of prefixes for removing
prefix = ['Time from Start to Finish (seconds)','Q8', 'Q11','Q12_',
          'Q12_OTHER', 'Q13','Q17_','Q18_', 'Q19_','Q20', 'Q21', 'Q22',
          'Q23', 'Q24_', 'Q26','Q27_', 'Q28','Q29_', 'Q30_', 'Q30_','Q31_',
          'Q32_','Q33', 'Q33_', 'Q33_B','Q34_', 'Q35_', 'Q35', 'Q36_', 'Q37_',
          'Q38_A', 'Q39_','Q42_', 'Q27_','Q29_', 'Q30_', 'Q38_B' 
          ]

# find columns with prefix
columns_to_drop = [
     col for col in df21.columns if any(col.startswith(p) for p in prefix)
     ]
# print(columns_to_drop)

# remove columns from dataframe
df21 = df21.drop(columns=columns_to_drop)

# new dataframe
df21.shape

df21.columns

for col in df21.columns: #columns in new dataframe
     print(col)
```

### Correction of headers

Combination of two rows for preparation of main header

```python
#original header
original_cols = df21.columns.tolist()
line_0 = df21.iloc[0].tolist()

# new header combination with first one
new_header = [f"{col}_{val}" for col, val in zip(original_cols, line_0)]

# setting new header and remove first row
df21.columns = new_header
df21 = df21.iloc[1:].reset_index(drop=True)
df21
```

### Splitting columns that contain multiple tools or their versions
